```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

# Categorical Feature Optimization

```python
files = glob.glob('categorical_features/raw/XGBoost/*label.pkl')
dfs = []
for f in files:
    df = pd.read_pickle(f)
    df['date'] = f.split('/')[3].split('_')[2].split('.')[0]
    df['encoding'] = 'label'
    dfs.append(df)
label_df = pd.concat(dfs)
label_df['r2'] = label_df.r2.apply(lambda x: x[0] if len(x) > 0 else np.nan)
label_df['rmse'] = label_df.rmse.apply(lambda x: x[0] if len(x) > 0 else np.nan)
label_df.sort_values(['date','features'], inplace=True)
#label_df.to_pickle('categorical_features_optimization_label_XGBoost.pkl')
label_df
```

```python
files = glob.glob('categorical_features/raw/XGBoost/*onehot.pkl')
dfs = []
for f in files:
    df = pd.read_pickle(f)
    df['date'] = f.split('/')[3].split('_')[2].split('.')[0]
    df['encoding'] = 'onehot'
    dfs.append(df)
onehot_df = pd.concat(dfs)
onehot_df['r2'] = onehot_df.r2.apply(lambda x: x[0] if len(x) > 0 else np.nan)
onehot_df['rmse'] = onehot_df.rmse.apply(lambda x: x[0] if len(x) > 0 else np.nan)
onehot_df.sort_values(['date','features'], inplace=True)
#onehot_df.to_pickle('categorical_features_optimization_onehot_XGBoost.pkl')
onehot_df
```

```python
files = glob.glob('categorical_features/raw/XGBoost/*target.pkl')
dfs = []
for f in files:
    df = pd.read_pickle(f)
    df['date'] = f.split('/')[3].split('_')[2].split('.')[0]
    df['encoding'] = 'target'
    dfs.append(df)
target_df = pd.concat(dfs)
target_df['r2'] = target_df.r2.apply(lambda x: x[0] if len(x) > 0 else np.nan)
target_df['rmse'] = target_df.rmse.apply(lambda x: x[0] if len(x) > 0 else np.nan)
target_df.sort_values(['date','features'], inplace=True)
#target_df.to_pickle('categorical_features_optimization_target_XGBoost.pkl')
target_df
```

```python
idx_list_label  = list()
mean_list_label = list()
for idx, grp in label_df.groupby('features'):
    #print(idx, grp['rmse'].mean())
    idx_list_label.append(idx)
    mean_list_label.append(grp['rmse'].mean())

sorted_lists = sorted(zip(mean_list_label, idx_list_label))
sorted_mean_label = [x[0] for x in sorted_lists]
sorted_idx_label = [x[1] for x in sorted_lists]

for i,idx in enumerate(sorted_idx_label):
    sorted_idx_label[i] = str(idx)
```

```python
new_ylabels = ['User & Partition', 'User', 'User, Partition & QOS', 'User & QOS', 'User & Name']
colors = [(255/255,165/255,0/255),(251/255,73/255,135/255),(82/255,218/255,147/255),(172/255,142/255,214/255),(113/255,180/255,234/255)]
fig, ax = plt.subplots(figsize=[12,6])
for i in range(5):
    ax.barh(new_ylabels[i], np.array(sorted_mean_label[i]) / 3600, color=colors[i])
ax.set_xlabel('Average RMSE', fontsize=12)
ax.set_title('Error by Categorical Feature Combination: Label Encoding', fontsize=16)
ax.tick_params(axis='both',which='major',labelsize=10,length=0,pad=5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
plt.xlim([7,8])
plt.tight_layout()
#plt.savefig('error_by_categorical_feature_combination_label_encoding.jpg')
```

```python
idx_list_onehot = list()
mean_list_onehot = list()
for idx, grp in onehot_df.groupby('features'):
    idx_list_onehot.append(idx)
    mean_list_onehot.append(grp['rmse'].mean())
    
sorted_lists = sorted(zip(mean_list_onehot, idx_list_onehot))
sorted_mean_onehot = [x[0] for x in sorted_lists]
sorted_idx_onehot = [x[1] for x in sorted_lists]

for i,idx in enumerate(sorted_idx_onehot):
    sorted_idx_onehot[i] = str(idx)
```

```python
new_ylabels = ['Account, Directory & Job Name', 'User, Account, Directory & Job Name', 'Directory & Job Name', 'User, Account & Name', 'Account & Name']
colors = [(255/255,165/255,0/255),(251/255,73/255,135/255),(82/255,218/255,147/255),(172/255,142/255,214/255),(113/255,180/255,234/255)]
fig, ax = plt.subplots(figsize=[12,6])
for i in range(5):
    ax.barh(new_ylabels[i], np.array(sorted_mean_onehot[i]) / 3600, color=colors[i])
ax.set_xlabel('Average RMSE', fontsize=12)
ax.set_title('Error by Categorical Feature Combination: One-Hot Encoding', fontsize=16)
ax.tick_params(axis='both',which='major',labelsize=10,length=0,pad=5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
plt.xlim([7,8])
plt.tight_layout()
#plt.savefig('error_by_categorical_feature_combination_onehot_encoding.jpg')
```

# Numerical Feature Optimization

```python
files = glob.glob('numerical_features/raw/XGBoost/numerical*')
```

```python
dfs = []
for f in files:
    df = pd.read_pickle(f)
    df['date'] = f.split('/')[3].split('_')[2].split('.')[0]
    dfs.append(df)
numerical_df = pd.concat(dfs)
numerical_df['r2'] = numerical_df.r2.apply(lambda x: x[0] if len(x) > 0 else np.nan)
numerical_df['rmse'] = numerical_df.rmse.apply(lambda x: x[0] if len(x) > 0 else np.nan)
numerical_df.sort_values(['date','features'], inplace=True)
#numerical_df.to_pickle('numerical_features_optimization_XGBoost.pkl')
numerical_df
```

```python
idx_numerical = list()
mean_numerical = list()
for idx, grp in numerical_df.groupby('features'):
    idx_numerical.append(idx)
    mean_numerical.append(grp['rmse'].mean())

for i,idx in enumerate(idx_numerical):
    idx_numerical[i] = str(idx)

sorted_lists = sorted(zip(mean_numerical, idx_numerical))
sorted_mean_numerical = [x[0] for x in sorted_lists]
sorted_idx_numerical = [x[1] for x in sorted_lists]
```

```python
def replace_in_string(s):
    s = s.replace('P', 'Processors')
    s = s.replace('G', 'GPUs')
    s = s.replace('M', 'Mem')
    s = s.replace('N', 'Nodes')
    return s
```

```python
new_ylabels=['WC','WC / G','WC / G / M','WC / M','WC / N','WC / G / N','WC / G / M / N','WC / M / N',\
             'WC / N / P','WC / N / G / P', 'WC / G / P', 'WC / P',\
             'WC / N / G / P / M','WC / N / P / M','WC / G / P / M','WC / P / M']
new_ylabels.reverse()

for i,s in enumerate(new_ylabels):
    new_ylabels[i] = replace_in_string(s)

sorted_mean_numerical_seconds = np.array(sorted_mean_numerical)/3600

colors = [(251/255,73/255,135/255),(82/255,218/255,147/255),(172/255,142/255,214/255),(113/255,180/255,234/255)]
groups = ['Includes only GPUs or Memory','Includes Nodes','Includes Processors','Includes Processors & Memory']
legend_handles = [mpatches.Patch(color=colors[3-i], label=groups[i]) for i in range(4)]
fig, ax = plt.subplots(figsize=[12,6])
y_positions = list(np.arange(0,4.25,.25))

print(sorted_mean_numerical)
ax.barh(y_positions[:4], sorted_mean_numerical_seconds[:4], color=colors[0], height=0.2)
ax.barh(y_positions[4:8], sorted_mean_numerical_seconds[4:8], color=colors[1], height=0.2)
ax.barh(y_positions[8:12], sorted_mean_numerical_seconds[8:12], color=colors[2], height=0.2)
ax.barh(y_positions[12:16], sorted_mean_numerical_seconds[12:16], color=colors[3], height=0.2)

for i in range(16):
    plt.text(sorted_mean_numerical_seconds[i] + .01, y_positions[i], new_ylabels[i], va='center')

ax.set_xlabel('Average RMSE', fontsize=12)
ax.set_yticklabels([])
ax.set_title('Error by Categorical Feature Combination: One-Hot, Target, & Label Encoding', fontsize=16)
ax.tick_params(axis='both',which='major',labelsize=10,length=0,pad=5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1)
ax.spines['left'].set_linewidth(1)
ax.legend(handles=legend_handles, loc='lower right', fontsize=10)
plt.xlim([7,9.5])
plt.tight_layout()
plt.savefig('error_by_numerical_feature_combinations.jpg')
```

## TFIDF Model

```python
files = glob.glob('training_window/raw/tfidf/training*')
```

```python
def convert_values(x):
    if not isinstance(x, list):
        return x
    else:
        if len(x) > 0:
            return x[0]
        else:
            return np.nan
```

```python
dfs = []
for f in files:
    df = pd.read_pickle(f)
    df['date'] = f.split('/')[3].split('_')[3].split('.')[0]
    dfs.append(df)
training_tfidf_df = pd.concat(dfs)
training_tfidf_df['r2'] = training_tfidf_df.r2.apply(convert_values)
training_tfidf_df['rmse'] = training_tfidf_df.rmse.apply(convert_values)
training_tfidf_df.sort_values(['date','training_window'], inplace=True)
training_tfidf_df.to_pickle('training_window_optimization_TFIDF.pkl')
training_tfidf_df
```

```python
idx_tfidf_training = list()
mean_tfidf_training = list()
for idx, grp in training_tfidf_df.groupby('training_window'):
    idx_tfidf_training.append(idx)
    mean_tfidf_training.append(grp['rmse'].mean())
```

```python
fig = plt.figure(figsize=(6,4))
plt.scatter(idx_tfidf_training, np.array(mean_tfidf_training) / 3600)
plt.title('Optimizing Training Window: TFIDF')
plt.xlabel('Training Window (days)')
plt.ylabel('RMSE (hours)')
plt.ylim([8,18])
#plt.savefig('optimize_training_window_TFIDF.jpg', dpi=300)
```
